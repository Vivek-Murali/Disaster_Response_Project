### 2. Merge both datasets.
- Merge the messages and categories datasets using the common id
- Assign this combined dataset to `df`, which will be cleaned in the following steps

```python
# inner join on the Field ID
df = pd.merge(messages, categories, on='id', how='outer')
df.head()
```

```python
categories = df['categories'].str.split(";", expand=True)
categories.head()
```

```python
row = categories.iloc[0]

# use this row to extract a list of new column names for categories.
# one way is to apply a lambda function that takes everything 
# up to the second to last character of each string with slicing
category_colnames = list(map(lambda x: x.split("-")[0], categories.iloc[0].values.tolist()))
print(category_colnames)
```

```python
#renaming the column names
categories.columns = category_colnames
categories.head()
```

```python
for column in categories:
    # set each value to be the last character of the string
    categories[column] = categories[column].str[-1]
    
    # convert column from string to numeric
    categories[column] = categories[column].apply(pd.to_numeric)
categories.head()
```

```python
# Concating bpth Dataframe with Imputed Values
df = pd.concat([df, categories], axis=1)
df.head()
```

```python
## Counting duplicates
count_duplicates = 0
for item in df.duplicated():
    if item == True:
        count_duplicates += 1
print("Number of duplicates in the dataset: {}".format(count_duplicates))
```

```python
df = df.drop_duplicates(keep='first')
```

```python
engine = create_engine('sqlite:///disaster_response.db')
df.to_sql('messages1', engine, index=False)
```
